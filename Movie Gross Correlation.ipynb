```python
#Import Libraries 

import pandas as pd
import seaborn as sns
import numpy as np

import matplotlib
import matplotlib.pyplot as plt
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8) # Configuration of the plots that will be created
```

```python
# Read in the data 
df = pd.read_csv(r'C:\Users\MoazA\Downloads\movies.csv')
```

```python
# Look at the first 5 rows of the dataset

df.head()
```

```python
# Check for missing data

for col in df.columns:
    missing_val_percent = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_val_percent))
    
# Remove rows with NULL values
df = df.dropna()

for col in df.columns:
    missing_val_percent = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, missing_val_percent))
```

```python
# Check data type for our columns

df.dtypes
```

```python
# Convert votes, budget, gross, and runtime data type to integers
#(decimals are not needed in these columns, looks cleaner)

df['votes'] = df['votes'].astype('int64')
df['budget'] = df['budget'].astype('int64')
df['gross'] = df['gross'].astype('int64')
df['runtime'] = df['runtime'].astype('int64')

```

```python
# Check data type again

df.dtypes
```

```python
df.sort_values(by=['gross'], inplace = False, ascending = False)
```

```python
# Remove duplicates, if any exist

df.drop_duplicates()
```

```python
# Check for correlation
```

```python
# Scatter plot of Budget vs Gross

plt.scatter(x = df['budget'], y = df['gross'])
plt.title('Budget Vs. Gross Earnings')
plt.xlabel('Movie Budget')
plt.ylabel('Gross Earnings')

plt.show()
```

```python
# Regression Plot of Budget vs. Gross

sns.regplot(x = 'budget', y = 'gross', data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "green"})
plt.title('Regression Plot of Budget Vs. Gross Earnings')
plt.xlabel('Movie Budget')
plt.ylabel('Gross Earnings')
# Format y-axis ticks
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:} Billion'.format(x*1e-9) for x in current_values])
# Format x-axis ticks
current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:} Million'.format(x*1e-6) for x in current_values])

plt.show()
```

```python
# Look for correlation on numerical columns

df.corr() #Pearson Correlation
```

```python
# Shows high correlation between Gross and Budget
```

```python
# Visualize the numerical correlation with a heatmap

correlation_matrix = df.corr()

sns.heatmap(correlation_matrix, annot = True)
plt.title('Correlation Metric for Numerical Features')
plt.xlabel('Movie Feature')
plt.ylabel('Movie Feature')

plt.show()
```

```python
# Count of IMDB users votes and Gross has a good correlation of 0.61

# Regression Plot of Votes vs. Gross

sns.regplot(x = 'votes', y = 'gross', data = df, scatter_kws = {"color" : "red"}, line_kws = {"color" : "green"})
plt.title('Regression Plot of Budget Vs. Gross Earnings')
plt.xlabel('Movie Vote')
plt.ylabel('Gross Earnings')
# Format y-axis ticks
current_values = plt.gca().get_yticks()
plt.gca().set_yticklabels(['{:} Billion'.format(x*1e-9) for x in current_values])
# Format x-axis ticks
current_values = plt.gca().get_xticks()
plt.gca().set_xticklabels(['{:} Million'.format(x*1e-6) for x in current_values])

plt.show()
```
